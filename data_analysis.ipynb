## Safety Features

Start_/Stop_Button, Handbrake, Hill_Assist, Child_Safety_Lock, Number_of_Airbags, Rain_Sensing_Wipers, Airbags, Parking_Assistance are the features that play a major role in safety of people in the car. These features may help help in preventing accidents,life loses and minimizing car damage

```python
sns.boxplot(x=df1['Start_/_Stop_Button'], y=df1['Ex-Showroom_Price'])
plt.xticks(rotation='vertical')
plt.show()
# cars with Start/stop button are more expensive
```

```python
# df1['Start_/_Stop_Button'].value_counts().plot(kind='bar')
sns.countplot(x='Start_/_Stop_Button', data=df1, palette='ocean')
plt.show()
# but almost 600 cars donot have start/stop button
```

```python
sns.barplot(x='Make', y='Ex-Showroom_Price', hue='Start_/_Stop_Button',data=df1, palette='pink')
plt.xticks(rotation='vertical')
plt.show()
# datsun, bajaj, icml, force, skoda and mg manufacturers donot make cars with start/stop button 
```

```python
df1['Parking_Assistance']=df1['Parking_Assistance'].replace('Yes', np.nan)
df1['Parking_Assistance']=df1['Parking_Assistance'].fillna(method='ffill')
df1['Parking_Assistance']=df1['Parking_Assistance'].replace('Rear sensors, Rear sensors with camera','Rear sensors with camera, Rear sensors')
df1['Parking_Assistance']=df1['Parking_Assistance'].replace('Rear sensors, Front sensors', 'Front sensors, Rear sensors')
sns.boxplot(x=df1['Parking_Assistance'], y=df1['Ex-Showroom_Price'])
plt.xticks(rotation='vertical')
plt.show()
# we cannot comment upon parking assistance
```

```python
sns.countplot(x='Parking_Assistance', data=df1, palette='ocean_r')
plt.xticks(rotation='vertical')
plt.show()
# many cars donot have parking assistance and if it is presnt than either they have rear sensors or rear sensors with camera
```

```python
sns.boxplot(x=df1['Number_of_Airbags'], y=df1['Ex-Showroom_Price'])
plt.xticks(rotation='vertical')
plt.show()
# as number of airbags increases, car price definitely increases
```
